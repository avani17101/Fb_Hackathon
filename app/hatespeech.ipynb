```python
path = '/home/lenovo/Fb_Hackathon/app/Friend_Indeed'
import os
import json
import pandas as pd

def create_fileEntity(entityName):
    path = '/home/lenovo/Fb_Hackathon/app/Friend_Indeed'
    filepath = os.path.join(path+ '/entities/'+entityName+'.json')
    if not os.path.exists(path):
        os.makedirs(path)
    f = open(filepath, "a")
    return (f,path)
```
